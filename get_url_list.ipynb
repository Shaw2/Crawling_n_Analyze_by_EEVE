```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_post_urls(board_url):
    """사진 게시판에서 게시물의 URL 리스트를 추출하는 함수"""
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(board_url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        # print("soup : ", soup)
        # post_links = [a["href"] for a in soup.find_all("a", class_="post-link")]  # 게시물 링크 추출
        post_links = []  # 게시물 링크를 저장할 빈 리스트 생성

        for div_block in soup.find_all("div", class_="img-box"):  # class가 "post-link"인 모든 <a> 태그 찾기
            a_tag = div_block.find("a")
            
            # print("type(href) : ", type(a_tag))
            # print("href : ", a_tag)
            
            href = a_tag["href"]# 각 <a> 태그의 href 속성 값을 가져오기
            real_href = href[2:]

            post_links.append(real_href)  # 가져온 href 값을 리스트에 추가        
        
        print("len(post_links) : ", len(post_links))
        print("post_links : ", post_links)
        
        return post_links
    else:
        return []

def extract_urls_from_post(post_url):
    """특정 게시물 내부에서 URL을 추출하는 함수"""
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(post_url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        # extracted_urls = [a["href"] for a in soup.find_all("a", class_="target-url")]  # 특정 부분에서 URL 추출
        
        extracted_urls = []
        for p_block in soup.find_all("p", class_="url"):  # class가 "post-link"인 모든 <a> 태그 찾기
            a_tag = p_block.find("a")
            site_url = a_tag["href"]
            
            extracted_urls.append(site_url)
            print("extracted_urls : ", extracted_urls)
        return extracted_urls
    else:
        return []
    
def make_csv(site_lists):
    print("len(Check site_lists) : ", len(site_lists))
    print("Check site_lists : ", site_lists)
    
    df = pd.DataFrame(site_lists, index=None)
    
    print("Dataframe : ", df)
    
    df.to_csv("C:/Users/Dolphinnn/VscodeProjects/Crawling_n_Analyze_by_EEVE/data/Enterprise_url_from_GDWEB.csv", header=False)
```

```python
if __name__ == "__main__":
    
    # 메인에서 '기업소개' 카테고리를 누르면 37개가 뜨는데, 
    # 20번째까지가 2018년도 4월 사이트 까지여서, 사이트 생존 확률 및 퀄리티는 위해서
    # 20번째 탭까지만 크롤링 하기로 하였음
    # 1개의 탭에 60개의 사이트가 들어가니까, 최대 1200개의 기업소개 사이트 HTML을 볼 수 있음음
    
    page_number = 20
    
    all_extracted_urls = []
    
    for current_page_number in range(1, page_number+1):
        print(f"current_page_number {current_page_number} is processing...", )
        
        board_url = f"https://www.gdweb.co.kr/sub/list.asp?displayrow=60&Txt_key=all&Txt_word=&Txt_agnumber=&Txt_fgbn=5&Txt_bcode1=022710001&Txt_gbflag=&Txt_bcode2=&Txt_bcode3=&Txt_bcode4=&Txt_bcode5=&Page={current_page_number}"   # 게시판 URL 
        post_urls = get_post_urls(board_url)
        
        for post_url in post_urls:
            full_post_url = f"https://www.gdweb.co.kr/sub/{post_url}"  # 전체 URL 구성
            extracted_urls = extract_urls_from_post(full_post_url)
            all_extracted_urls.extend(extracted_urls)
        
    print("추출된 URL 리스트 갯 수 : ", len(all_extracted_urls))
    print("추출된 URL 리스트 : ", all_extracted_urls)
    
    make_csv(all_extracted_urls)
    
```
